# Método de la trasnformada Inverasa

Se va a elaborar un código que implemente el método de la transformada inversa para generar muestras de una distribución exponencial.

## Funciones

```python
import numpy as np
import matplotlib.pyplot as plt

#FDP de la distribución exponencial
def exponencial_pdf(x, lmbda):
    return lmbda * np.exp(-lmbda * x)

#CDF de la distribución exponencial
def exponencial_cdf(x, lmbda):
    return 1 - np.exp(-lmbda * x)

#Método de la transformada inversa
def transformada_inversa(lmbda, t_muestra):
    U = np.random.uniform(0, 1, t_muestra)
    return -np.log(1 - U) / lmbda
```

## Parámetros

```python
#Agrwgamos el parámetro lambda de la distribución exponencial
lmbda = 1.0

#Tamaño de la muestra
t_muestra = 10000
muestras = transformada_inversa(lmbda, t_muestra)

#Valores x para graficar
x_values = np.linspace(0, 10, 1000)
```

## Gráficas

```python
#FDP GRÁFICA
plt.figure(figsize=(12, 5))

plt.subplot(1, 3, 1)
plt.plot(x_values, exponencial_pdf(x_values, lmbda), label='fdp', color='blue')
plt.title('FDP GRÁFICA')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

#CDF GRÁFICA
plt.subplot(1, 3, 2)
plt.plot(x_values, exponencial_cdf(x_values, lmbda), label='CDF', color='green')
plt.title('FDA GRÁFICA')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.legend()

#Histograma de las muestras
plt.subplot(1, 3, 3)
plt.hist(muestras, bins=50, density=True, alpha=0.6, color='orange')
plt.plot(x_values, exponencial_pdf(x_values, lmbda), label='fdp', color='blue')
plt.title('Histograma de las muestras')
plt.xlabel('x')
plt.ylabel('Frecuencia')
plt.legend()

plt.tight_layout()
plt.show()
```

##Documentación

1. **Función de Densidad de Probabilidad (fdp):** La función `exponencial_pdf` calcula la fdp de una distribución exponencial con el parámetro \(\lambda\).
2. **Función de Distribución Acumulativa (CDF):** La función `exponencial_cdf` calcula la CDF de una distribución exponencial con el parámetro \(\lambda\).
3. **Método de la Transformada Inversa:** La función `transfurmada_inversa` genera muestras de una distribución exponencial utilizando el método de la transformada inversa. Primero, se generan números uniformes \(U\) entre 0 y 1, y luego se aplican la transformada inversa para obtener las muestras exponenciales.
4. **Generación de Muetras:** Se generan 10,000 muestras utilizando el método de la transformada inversa.
5. **Graficación:**
